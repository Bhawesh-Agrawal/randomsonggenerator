```python
def markov_chain(song):
    word = song.split()
    for i in range(len(word)-1):
        cur_word = word[i].lower()
        next_word = word[i+1].lower()

        if cur_word not in df:
            df[cur_word] = {}

        if next_word not in df[cur_word]:
            df[cur_word][next_word] = 0

        df[cur_word][next_word] += 1

    for cur_word, transition in df.items():
        total = float(sum(df[cur_word].values()))
        for next_word in transition:
            transition[next_word] /= total       
```

```python
def predict_next_word(df, cur_word):
    if cur_word in df:
        next_words = list(df[cur_word].keys())
        probabilities = list(df[cur_word].values())
        return np.random.choice(next_words, p=probabilities)
    else:
        return None
```
